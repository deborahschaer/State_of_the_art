```python
from data import load_papers 
import numpy as np
import matplotlib.pyplot as plt
```

```python
filename = 'write_summary.txt_bert.txt'
filename = 'write_summary_complex100.txt_selfeval1.txt'
with open('results/'+filename, 'r') as file:
    lines = file.readlines()

# Step 2: Extract the values
values = []
for line in lines:
    # Split each line by whitespace and convert to float
    values.extend([float(x) for x in line.split()])

# Step 3: Create a histogram
plt.hist(values, bins=10, edgecolor='black')
plt.title('Histogram of Values')
plt.xlabel('Value')
plt.ylabel('Frequency')
plt.show()
```
