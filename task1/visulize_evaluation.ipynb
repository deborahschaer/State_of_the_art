```python
from data import load_papers 
import numpy as np
import matplotlib.pyplot as plt
```

```python
filename = 'write_summary100.txt_bert.txt'
with open('results/'+filename, 'r') as file:
    lines = file.readlines()

# Step 2: Extract the values
P = []
R = []
F1_values = []
for i in range(100):
    P.append(float(lines[i].split()[0]))
    R.append(float(lines[i].split()[1]))
    F1_values.append(float(lines[i].split()[2])) 
print(len(F1_values))
plt.boxplot([P,R,F1_values])
plt.title('Boxplot using matplotlib')
plt.show()
```

```python
filename = 'write_summary100.txt_selfeval1.txt'
with open('results/'+filename, 'r') as file:
    lines = file.readlines()

# Step 2: Extract the values
self1 = []
for i in range(100):
    self1.append(float(lines[i]))

filename = 'write_summary100.txt_selfeval2.txt'
with open('results/'+filename, 'r') as file:
    lines = file.readlines()

# Step 2: Extract the values
self2 = []
for i in range(100):
    self2.append(float(lines[i]))


filename = 'write_summary100.txt_rouge.txt'
with open('results/'+filename, 'r') as file:
    lines = file.readlines()

# Step 2: Extract the values
rouge = []
for i in range(100):
    rouge.append(float(lines[i].split()[2]))

plt.boxplot([rouge,F1_values,self1,self2])
plt.title('Scores')
plt.xticks([1, 2, 3,4], ['Rouge','Bert', 'self1', 'self2'], rotation=45)
plt.show()
```

```python
filename = 'write_summary_complex100.txt_selfeval1.txt'
with open('results/'+filename, 'r') as file:
    lines = file.readlines()


# Step 2: Extract the values
self1_c = []
for i in range(100):
    self1_c.append(float(lines[i]))

filename = 'write_summary_complex100.txt_selfeval2.txt'
with open('results/'+filename, 'r') as file:
    lines = file.readlines()

# Step 2: Extract the values
self2_c = []
for i in range(100):
    self2_c.append(float(lines[i]))


filename = 'write_summary_complex100.txt_rouge.txt'
with open('results/'+filename, 'r') as file:
    lines = file.readlines()

# Step 2: Extract the values
rouge_c = []
for i in range(100):
    rouge_c.append(float(lines[i].split()[2]))

filename = 'write_summary_complex100.txt_bert.txt'
with open('results/'+filename, 'r') as file:
    lines = file.readlines()

# Step 2: Extract the values
bert_c = []
for i in range(100):
    bert_c.append(float(lines[i].split()[2]))
```

```python
import matplotlib.pyplot as plt
import numpy as np

# Example data
np.random.seed(10)
scores_data1 = [rouge,F1_values,self1,self2]
scores_data2 = [rouge_c,bert_c,self1_c,self2_c]

# Combine the two sets of data
combined_data = [scores_data1, scores_data2]
labels = ['Rouge', 'Bertscore', 'Self 1', 'Self 2']

# Create a boxplot
fig, ax = plt.subplots()

# Offset positions for the two sets of boxplots
positions_data1 = np.array(range(len(scores_data1))) * 2.0 - 0.3
positions_data2 = np.array(range(len(scores_data2))) * 2.0 + 0.3

# Plot boxplots for both datasets
box1 = ax.boxplot(scores_data1, positions=positions_data1, widths=0.4, patch_artist=True, boxprops=dict(facecolor="lightblue"))
box2 = ax.boxplot(scores_data2, positions=positions_data2, widths=0.4, patch_artist=True, boxprops=dict(facecolor="lightgreen"))

# Set x-axis labels
ax.set_xticks(np.arange(len(scores_data1)) * 2.0)
ax.set_xticklabels(labels)

# Add a legend
ax.legend([box1["boxes"][0], box2["boxes"][0]], ["Simple Prompt", "Complex Prompt"], loc='upper left')

plt.show()
```
